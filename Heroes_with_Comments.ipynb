## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



Here I used the same function agg, it in simple words let us to execute multiple operations per column in the originated group, again I used the size function to know jow many items or records are in the group, the colum price will be summed to obtain a total.
After that I order the dataframe by two columns in descending order

Finally the important ppint is the join. Basically, for example if I have a dataframe A with columns X Y Z and the dataframe with columns X M A, the join will create a dataframe with columns X Y Z M A, as we can see the pivot or colum that will let know python how to arrange the records is the X. In our case the Item ID is the pivot column

```python
groupPopular = df_purchase_data2.groupby('Item ID').agg({'Item ID':'size',                                                       
                                                       'Price':'sum'})
groupPopular.columns= ['Purchase Count','Total Purchase Value']
groupPopular = groupPopular.sort_values(['Purchase Count','Item ID'], ascending = False) 

groupPopular = groupPopular.join(df_purchase_data2[['Item Name', 'Price']], on='Item ID')

groupPopular.set_index(['Item Name'], inplace = True, append = True)

groupPopular = groupPopular[['Purchase Count','Price','Total Purchase Value']]

groupPopular.head()
```
